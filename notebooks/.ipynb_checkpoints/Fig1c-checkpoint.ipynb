```python
df['Date'] =pd.to_datetime(df.Date)
df['month_year'] = pd.to_datetime(df['Date']).dt.to_period('M')
```

```python
# keep months only
# one study should count as one
# log scale?
from scipy.interpolate import interpolate
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Helvetica"
plt.rc('ytick', labelsize=5.5) 
plt.rc('xtick', labelsize=4) 
plt.rc('axes', titlesize=5.5)
plt.rc('axes', labelsize=5.5)

# plt.plot(df1['Date'], df1['cumulative'],linewidth=7.0)
plt.plot(df['ym'], df.index,linewidth=1.0,color="#EE7733",marker='s',markersize=2)
plt.plot(df1['ym'], df1.index,linewidth=1.0,color="#44AA99",marker='^',markersize=2)
# plt.gcf().set_size_inches(10, 10)
plt.gcf().set_size_inches(2.416, 2.458)
# plt.set_xlabels(my_xticks)
labels=["Genomes Sequenced","WWTP Samples"]
plt.legend(labels,loc="lower right", prop={'size': 4.5},edgecolor="white")
# plt.yscale('log') 
plt.xlabel("Date")
plt.tight_layout()

plt.ylabel("Total number of studies")
# plt.grid(color="#DDDDDD")
positions = ('2019-12','2020-01','2020-02','2020-03','2020-04','2020-05','2020-06','2020-07','2020-08','2020-09','2020-10')
labels1 = ('Dec','Jan-20','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct')
plt.xticks(positions, labels1)
# plt.show()
plt.grid(linewidth=0.3,color="#DDDDDD")
plt.savefig("combined_samples_vs_time.png",dpi=300)
plt.savefig("combined_samples_vs_time.pdf",dpi=300)
```
