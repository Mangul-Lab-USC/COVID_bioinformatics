```python
import matplotlib.pyplot as plt

fig =plt.figure(figsize = (1,1))
ax11 = fig.add_subplot(111)
# Data to plot
sizes = [215, 130, 245, 2000]
labels = 'Python', 'C++', 'Ruby', 'Java'
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
w,l,p = ax11.pie(sizes,  labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=140, pctdistance=0.65, radius = 1, explode=explode)
[t.set_rotation(0) for t in p]
[t.set_fontsize(25) for t in p]
[t.set_fontsize(25) for t in l]
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = 'black'
plt.rcParams['axes.labelcolor']= '#909090'
plt.rcParams['xtick.color'] = '#909090'
plt.rcParams['ytick.color'] = '#909090'
plt.rcParams['font.size']=14
color_palette_list = ['#4477AA', '#66CCEE', '#BBBBBB']
labels =['Illumina','Oxford Nanopore','Others']

explode=(0,0,0)
sizes=[ 211258,33782,665]
patches, texts, autotexts  =  ax.pie(sizes, explode=explode, labels=labels,  
       colors=color_palette_list[0:3], 
        autopct='%1.1f%%', 
       shadow=False, startangle=0,   
       pctdistance=1.2,labeldistance=1.5)
# [text.set_color('red') for text in texts]
texts[0].set_color('#4477AA')
texts[1].set_color('#66CCEE')
texts[2].set_color('#BBBBBB')
# autotexts[0].set_color('#4477AA')
# autotexts[1].set_color('#66CCEE')
# autotexts[2].set_color('#BBBBBB')
# [autotext.set_color('black') for autotext in autotexts]
plt.title("Sequencing technologies in SRA database (246679 submissions)")
# ax.axis('equal')
# ax.set_title("Elephant in the Valley Survey Respondent Make-up")
ax.legend(frameon=False, bbox_to_anchor=(1.9,0.9))
plt.savefig('pie_SRA.png', dpi=300, bbox_inches = "tight")
plt.savefig('pie_SRA.pdf', dpi=300, bbox_inches = "tight")
```
